ノイズののったロジスティック写像

$X_t= a X_{t-1} (1-X_{t-1}) + \epsilon$

$\epsilon = N(0, \sigma^2)$

```python
import matplotlib.pyplot as plt
xs = np.zeros(10)+1
xs[0] = 0.1
a = 3.4

def logistic_map(x,a):
    return a * x * (1 - x)

def reflect(x):
    if x > 1:
        return reflect(1 - (x - 1))
    elif x < 0:
        return reflect(-x)
    else:
        return x
    
for i in np.arange(len(xs)-1)+1:
    xs[i]  = reflect( logistic_map(xs[i-1], a) + np.random.normal())
plt.plot(np.arange(len(xs)),xs)
```

カオスの性質:初期値鋭敏性

近くにある点の起動がどれくらい離れていくか、はリヤプノフ指数（Lyapunov exponet)で特徴づけられる

ずれの初期値を$\delta x_0$とする。

$|\delta x_n|=|\delta x_0|e^{\lambda n}$

と表すとすると、

$e^{\lambda n}  = \lim_{\delta x_0 \rightarrow 0} |\frac{\delta x_n}{\delta x_0}| = |\frac{dx_n}{dx_0}|$

$\frac{dx_n}{dx_0} = \frac{dx_n}{dx_{n-1}}\frac{dx_{n-1}}{dx_{n-2}}\cdots\frac{dx_1}{dx_0}$

$\lambda = \lim_{n\rightarrow \infty } \frac{1}{n} \ln |f'(x_{n-1}) \cdots f'(x_0)|$

となる。

$L(f,x)=\lim_{n\rightarrow \infty} \frac{1}{n} \log | (f^n)' (x)| $

$(f(n))'$はn回の再帰の導関数。これは計算が大変。そこで公式

$(f^n)'(x) = f'(x_{n-1}) \cdots f'(x_1)f'(x_0)$

帰納法で証明できる。


ふたつの点を$x_0,x_0+\delta x_0$とする。

$x_n=f(x_{n-1})$

$x_{n} + \delta x_{n} = f(x_{n-1} + \delta x_{n-1}) = a (x_{n-1} + \delta x_{n-1}) (1- x_{n-1} + \delta x_{n-1})  = f(x_{n-1}) + a x_{n-1} \delta x_{n-1} + \delta x_{n-1} (1-x_{n-1})= $

リヤプノフ指数

$L=\lim_{n\rightarrow \infty} \frac{1}{n} \sum \ln (|J|)$

J : ヤコビアン要素
