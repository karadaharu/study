### 3.1 伊藤積分の定義
ノイズが加わった方程式を確率過程$W_{t}$を用いて、

$\frac{dX}{dt}=b(t,X_{t})+\sigma(t,X_{t})・W_{t} \:\:\:\:(3.1.2)$

と表現します。

$W_{t}$は少なくとも近似的には次の性質をもたなくてはいけません。

1. $t_{1}\neq t_{2}$なら$W_{t_{1}}$と$W_{t_{2}}$は独立
2. $\{W_{t}\}$は定常。すなわち、同時分布$\{W_{t_{1}+t},\cdots W_{t_{k}+t}\}$はtに依らない
3. すべてのtについて$E[W_{t}]=0$となる

しかし、1,2を満たす"適切な"確率過程は存在しません。実際、そのような確率過程は連続な経路をもちません。さらに、$E[W_{t}^{2}]=1$と仮定すれば、写像$(t,\omega)\mapsto W_{t}(\omega)$はσ-加法族${\cal B}\times {\cal F}$に関して可測ですありません。

しかし、$W_{t}$を一般化された確率過程として実現することは可能です（これをホワイトノイズ過程と呼びます）。ここでいう"一般化された"とは、確率過程が通常の確率過程のように$R^{0,\infty)}$上の確率測度として実現されるのではなく、緩増加超関数の空間$\cal S'$の上の確率測度として構築されることを意味します。

ゆえに、この種の構成は放棄しましょう。むしろ、(3.1.2)を書き改めて、$W_{t}$を適当な確率過程で置き換える変形を考えましょう。このため、$0=t_{t}<\cdots < t_{m} = t$とおいて、(3.1.2)を差分方程式として次のように表してみましょう。

$X_{k+1}-K_{k}=b(t_{k},X_{k})\Delta t_{k} + \sigma(t_{k},X_{k})W_{k}\Delta t_{k}$

ただし、

$X_{j}=X(t_{j})$、$W_{k}=W_{t_{k}}$、$\Delta t_{k} = t_{k+1} - t_{k}$

さらに、$W_{k}$を用いた表現をやめ、新しい確率過程$\{V_{t}\}_{t \geq 0 }$を導入して、$W_{k} \Delta t_{k}$を$\Delta V_{k} = V_{t_{k+1}}-V_{t_{k}}$で置き換えましょう。$W_{t}$に関する仮定1,2,3は$V_{t}$が期待値0で定常かつ独立な増分をもつこと示唆しています。ところが、連続な経路をもつそのような確率過程はブラウン運動に限ります。したがって、$V_{t}=B_{t}$とおいて、(3.1.3)から次の関係式を得ます。

$X_{k}=X_{0}+\sum_{j=0}^{k-1}b(t_{j},X_{j})\Delta t_{j} + \sum_{j=0}^{k-1}\sigma (t_{j}, X_{j})\Delta B_{j}$

さて、$\Delta t_{j} \rightarrow 0$としたときに、何らかの意味で右辺が収束するといえるでしょうか？もしそうなら、

$X_{t} = X_{0} + \int^{t}_{0}b(t,X_{t})dt+"\int^{t}_{0}\sigma(t,X_{t})dB_{s}" $ (3.1.5) 

と表していいでしょうし、(3.1.2)は$X_{t}=X_{t}(\omega)$が(3.1.5)を満たす確率過程であることを意味していることと思えるでしょう。

なので、この章では、原点から出発する１次元ブラウン運動$B_{t}(\omega)$と十分に多くの関数$f:[0,\infty]\times \Omega \rightarrow {\bf R}$に対して、何らかの意味で

$"\int^{t}_{0}\sigma(t,X_{t})dB_{s}" $

が存在することを示しましょう。

$0\leq S \leq T$と$f(t,\omega)$が与えられたとします。このとき積分

$\int^{T}_{S}f(t,\omega)dB_{t}(\omega)$

を定義したいです。そのため、簡単な関数fに対する定義から始めて、近似法を用いて一般化するという手順を踏みましょう。まず、

$\phi(t,\omega)=\sum_{j\geq 0} e_{j}(\omega) \times {\cal X}_{[j2^{-n},(j+1)2^{-n})}(t)$

という形をしたΦをとります。ここで、${\cal X}$は定義関数（特性関数)であり、nは自然数です。

**定義関数**

集合Ωの部分集合Aについての定義関数とは、Ωの要素Pに対して

${\cal X}_{A}(P)=1 \: P \in A$  
${\cal X}_{A}(P)=0 \: P \notin A$  

となる関数${\cal X}_{A}$のことです。

$t_{k} = t_{k}^{(n)}=
\begin{cases}
    k2^{-n}, \:\: for S \leq k 2^{-n} \leq T\\
    S, \:\: for k 2^{-n} \leq S\\
    T, \:\: for k 2^{-n} \geq T\\
\end{cases}$

とおいて、上のような関数Φに対して、積分を次で定義します。

$\int_{S}^{T}\Phi (t , \omega) d B_{t} (\omega) = \sum_{j \geq 0} e_{j} (\omega) [B_{t_{j-1}}-B_{t_{j}}](\omega)$　(3.1.8)

しかし、次の例で見るように、$e_{j}(\omega)$に仮定をおかないと、この定義から近似法を用いて一般化するのは難しいです。

※以下Eは、原点を出発するブラウン運動$B_{t}$の法則$P^{0}$に関する期待値$E^{0}$を表すものとします。また、$P^{0}$を簡単にPと書きます。

例3.1.1  
$\phi_{1}(t,\omega)=\sum_{j\geq 0} B_{j2^{-n}}(\omega) \times {\cal X}_{[j2^{-n},(j+1)2^{-n})}(t)$  
$\phi_{2}(t,\omega)=\sum_{j\geq 0} B_{(j+1)2^{-n}}(\omega) \times {\cal X}_{[j2^{-n},(j+1)2^{-n})}(t)$

と定義します。$B_{t}$は独立な増分をもつから、

$E[\int^{T}_{0}\phi_{1}(t,\omega)dB_{t}(\omega)]=\sum_{j\geq 0}E[B_{t_{j}}(B_{t_{j+1}}-B_{t_{j}})]=0$

となります。しかし、(2.2.10)より

$E[\int^{T}_{0}\phi_{2}(t,\omega)dB_{t}(\omega)]=\sum_{j\geq 0}E[B_{t_{j+1}}(B_{t_{j+1}}-B_{t_{j}})]=\sum_{j\geq 0}E[(B_{t_{j+1}}-B_{t_{j}})^{2}]=T$
