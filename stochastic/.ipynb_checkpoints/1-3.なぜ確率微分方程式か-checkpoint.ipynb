## 1. 序
### 1.1 古典的微分方程式の確率論的近似
**問題1** 次のような人口変動のモデルを考えましょう。

$\frac{dN}{dt}=a(t)N(t), \: N(0) = N_{0}\:\:\:(1.1.1)$

このa(t)が完全には判明しない、つまりなんらかのランダムな環境の影響のために

a(t)=r(t)+"ノイズ"

という形をとることがあります。ここで、このノイズ項の正確な挙動は分からず、その確率分布だけ分かっています。この場合、どのように式(1.1.1)を解いたらいいでしょうか？

**問題2** 電気回路のある点での時刻tにおける電荷Q(t)は

$LQ''(t)+RQ'(t)+\frac{1}{C}Q(t)=F(t),\:Q(0)=Q_{0},Q'(0)=I_{0}$

という微分方程式を満たします。F(t)は時刻tにおける印加電圧です。このF(t)が確定的ではなく、

F(t) = G(t) + "ノイズ"

という形をしているかもしれません。このとき、電荷Q（ｔ）についてどのように解けばいいでしょうか？

### 1.2 フィルターの問題
**問題3** 例えば、問題2の解に関する情報を得るために、時刻s≦tにおいて、Q(s)の観測Z(s)を行ったとしましょう。しかし、正確に測定できず、

Z(s)=Q(s)+"ノイズ"          (1.2.1)

という観測がなされたとします。

フィルターの問題とは、「s≦tのとき、(1.2.1)のZ(s)に基づく、(1.1.2)を満たすQ(t)の最良の推定は何か？」という問題です。1961年にカルマンとブーシーがカルマン=ブーシー・フィルターとして知られるフィルター理論を証明しました。このフィルター理論によって、ノイズ項をもつ線形微分方程式に従うシステムの状態をノイズを含む一連の観測に基いて推定する方法が与えられました。

### 1.3 境界値問題への確率論的アプローチ
**問題4** もっとも著名な例は次のディリクレ問題の確率論的な解です。

「$R^{n}$の領域UとUの境界∂U上の連続関数fをとる。Uの閉包$\bar{U}$上で連続な関数$\tilde{f}$で

(i)∂U上$\tilde{f}=f$

(ii)U上$\tilde{f}$は調和である。すなわち、

$\Delta \tilde{f}(x) \equiv \sum_{i=1}^{n}\frac{\partial \tilde{f}}{\partial x^{2}_{i}}(x)=0, \forall x \in U$

1944年に角谷はブラウン運動を用いてこの解を構成できることを証明しました。

### 1.4 最適停止問題
**問題5**
売却する予定の資産を保有しているとしましょう。資産の時刻tにおける市場での価格$X_{t}$は問題1と同様にして、

$\frac{dX_{t}}{dt}=rX_{t}+\alpha X_{t}・\mbox{"ノイズ"}$

に従って変動します。そして減価率は既知の定数ρで与えられているとします。このとき、いつ売る決断を下せばいいでしょうか？

現在時刻tまでの$X_{t}$の挙動は分かっているとしましょう。しかし、システムがノイズを含むので、保有者は売却時にその選択が最良のものであるかどうかについて、もちろん確信をもてません。したがって、インフレーションを考慮にいれた期待効用を最大にする停止戦略を探さなければいけません。これが最適停止問題です。

### 1.5 確率制御
**問題6**

2種類の投資を考えます。

(1) 危険証券（たとえば株）　時間tにおける単位あたりの価格$p_{1}(t)$は問題1で考えたような確率微分方程式

$\frac{dp_{1}}{dt}=(a+\alpha ・\mbox{ノイズ})p_{1}$

に従います。

(2) 安全証券（たとえば債権）　時刻tにおける単位あたりの価格$p_{2}(t)$は次の方程式に従います

$\frac{dp_{2}}{dt}=bp_{2}$

資産の保有者は各時刻tに、資産額$X_{t}$のどれだけの割合($u_{t}$とする)を危険証券に投資するか決めます。

与えられた効用関数Uと満期期日Tに対し、最適なポートフォリオ$u_{t} \in [0,1]$を見出したいです。言い換えれば、

$\max_{0\leq u_{t} \leq 1} \{E[U(X_{t}^{(u)})]\}$

なる$u_{t}$、0≦t≦Tを見出したいです。

### 1.6 数理ファイナンス
時刻t=Tに特定の価格Kで危険証券を1単位買う権利が、時刻０に提示されたとしましょう。そのような権利をヨーロッパ型コールオプションと言います。保有者は、このオプションにどれほどの対価を喜んで支払うでしょうか？この問題が解決されたのは、F.BlackとM.Scholes(1973)が確率解析と均衡議論を用いたときでした。

これらの問題1~6について、これから考察していきます。

## 2. 準備
### 2.1 確率過程

**定義 2.1.4**  集合Tにより添字付けられた$(\Omega, {\cal F}, P)$上の($R^{n}$-値)確率変数の族

$\{X_{t}\}_{t \in T}$

を確率過程と呼びます。確率過程では、各$t \in T$に確率変数

$\omega \mapsto X_{t}(\omega), \: \omega \in \Omega$

が対応しています。視点を変えて、ωを固定し

$t \mapsto X_{t}(\omega), \: t \in T$

なる関数とみなすとき、経路と呼びます。

### 2.2 ブラウン運動
$x \in {\bf R}^{n}$を固定して、

$p(t,x,y)=(2 \pi t)^{-n/2} \times \exp(- \frac{|x-y|^{2}}{2t}), \: y \in {\bf R}^{n},t>0$

とします。（分散t、期待値yのガウス分布)　$0 \leq t_{1} \leq t_{2} \leq  \cdots \leq t_{k}$ に対し、${\bf R}^{nk}$上の測度$\nu_{t_{1},\cdots,t_{k}}$を

$\nu_{t_{1},\cdots,t_{k}}(F_{1}\times \cdots \times F_{k})=\int_{F_{1}\times \cdots \times F_{k}}p(t_{1},x,x_{1}) p(t_{2}-t_{1},x_{1},x_{2})\cdots p(t_{k}-t_{k-1},x_{k-1},x_{k})dx_{1}\cdots dx_{k}$

で定めます。

### 8.1 時間の連続化
ブラウン運動は、ランダム・ウォークの連続時間版です。

単純対称ランダム・ウォーク$S_{n}$の期待値と分散は、

$E(S_{n})=0$、$V(S_{n})=n$です。ここでランダム・ウォークの歩み幅を1から$\Delta r$に、時間幅を1から$\Delta t$に細かくします。すると、時刻tでの位置は

$D(t)=S_{[\frac{t}{\Delta t}]}・\Delta r$

です。いま、$\Delta t \rightarrow 0$、$\Delta r \rightarrow 0$としますが、$\Delta r = \sqrt{\Delta t}$の関係を保つものとします。$\Delta t = \frac{1}{n}$とすると(*この仮定の意味?*)、$\Delta r = \frac{1}{\sqrt{n}}$で、D(t)は、$Z_{1},Z_{2},\cdots$を確率1/2で±1をとる独立確率変数として、

$\frac{Z_{1}+Z_{2}+\cdots + Z_{[nt]}}{\sqrt{n}}=\sqrt{t}(\frac{Z_{1}+Z_{2}+\cdots+Z_{[nt]}}{\sqrt{nt}・1})$

と同じ確率分布になります。中心極限定理より、()部分の確率分布はN(0,1)に収束するので、全体は分散が$(\sqrt{t})^{2}$倍のN(0,t)に収束します。さらに、分散σ、平均μの2変数を決めて、

$\sigma ・ \sqrt{t} (\frac{Z_{1}+Z_{2}+\cdots+Z_{[nt]}}{\sqrt{nt}})+\mu t$


をつくると、$n\rightarrow \infty$のとき、これの確率分布が$N(\mu t,\sigma^{2}t)$となることはただちに分かります。すなわち、確率過程でX(t)でtを決めたときにその確率分布が$N(\mu t,\sigma^{2}t)$となるものが構成できます。  ここでμをドリフトパラメータ、σをゆらぎパラメータをいいます。 

この方法で進めると、次の正式定義を満たすブラウン運動を構成できます。

### 8.2 ブラウン運動の定義
**ブラウン運動**  
確率過程X(t)、t≧0が  
(i) $X(0)  \equiv 0$

さらにn+1個の時間の分点$0\leq t_{0} < t_{1} < \cdots < t_{n}$で作った変化分  
$X(t_{1})-X(t_{0}),X(t_{2})-X(t_{1}),\cdots ,X(t_{n})-X(t_{n-1})$  
が  
(ii) 独立  
で、かつ
(iii) それぞれ平均$\mu (t_{k}-t_{k-1})$、分散$\sigma^{2}(t_{k}-t_{k-1})$の正規分布  
$N(\mu (t_{k}-t_{k-1}), \sigma^{2}(t_{k}-t_{k-1}) )\: (k=1,2,\cdots n)$

に従う場合をブラウン運動といいます。  (ii)は「独立増分の仮定」、(iii)は「正規性の仮定」といいます。
