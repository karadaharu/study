### 8.1 時間の連続化
ブラウン運動は、ランダム・ウォークの連続時間版です。

単純対称ランダム・ウォーク$S_{n}$の期待値と分散は、

$E(S_{n})=0$、$V(S_{n})=n$です。ここでランダム・ウォークの歩み幅を1から$\Delta r$に、時間幅を1から$\Delta t$に細かくします。すると、時刻tでの位置は

$D(t)=S_{[\frac{t}{\Delta t}]}・\Delta r$

です。いま、$\Delta t \rightarrow 0$、$\Delta r \rightarrow 0$としますが、$\Delta r = \sqrt{\Delta t}$の関係を保つものとします。$\Delta t = \frac{1}{n}$とすると(*この仮定の意味?*)、$\Delta r = \frac{1}{\sqrt{n}}$で、D(t)は、$Z_{1},Z_{2},\cdots$を確率1/2で±1をとる独立確率変数として、

$\frac{Z_{1}+Z_{2}+\cdots + Z_{[nt]}}{\sqrt{n}}=\sqrt{t}(\frac{Z_{1}+Z_{2}+\cdots+Z_{[nt]}}{\sqrt{nt}・1})$

と同じ確率分布になります。中心極限定理より、()部分の確率分布はN(0,1)に収束するので、全体は分散が$(\sqrt{t})^{2}$倍のN(0,t)に収束します。さらに、分散σ、平均μの2変数を決めて、

$\sigma ・ \sqrt{t} (\frac{Z_{1}+Z_{2}+\cdots+Z_{[nt]}}{\sqrt{nt}})+\mu t$


をつくると、$n\rightarrow \infty$のとき、これの確率分布が$N(\mu t,\sigma^{2}t)$となることはただちに分かります。すなわち、確率過程でX(t)でtを決めたときにその確率分布が$N(\mu t,\sigma^{2}t)$となるものが構成できます。  ここでμをドリフトパラメータ、σをゆらぎパラメータをいいます。 

この方法で進めると、次の正式定義を満たすブラウン運動を構成できます。

### 8.2 ブラウン運動の定義
**ブラウン運動**  
確率過程X(t)、t≧0が  
(i) $X(0)  \equiv 0$

さらにn+1個の時間の分点$0\leq t_{0} < t_{1} < \cdots < t_{n}$で作った変化分  
$X(t_{1})-X(t_{0}),X(t_{2})-X(t_{1}),\cdots ,X(t_{n})-X(t_{n-1})$  
が  
(ii) 独立  
で、かつ
(iii) それぞれ平均$\mu (t_{k}-t_{k-1})$、分散$\sigma^{2}(t_{k}-t_{k-1})$の正規分布  
$N(\mu (t_{k}-t_{k-1}), \sigma^{2}(t_{k}-t_{k-1}) )\: (k=1,2,\cdots n)$

に従う場合をブラウン運動といいます。  (ii)は「独立増分の仮定」、(iii)は「正規性の仮定」といいます。

ブラウン運動をグラフに書くとき、ギザギザで途切れていないように書きますが、はたしてそれは正しいでしょうか？ブラウン運動について分かっている事項について解説します。

### 8.3 経路の連続性
**ブラウン運動の性質1**  
ブラウン運動の経路は確率1で連続である

証明は省略します。ブラウン運動が連続であることは、確率積分を考える上で、大きな拠り所となります。これをもとに、スティルチェス式積分を作り、確率積分を定義していくことになります。


**ブラウン運動の性質2**  
ブラウン運動の経路は確率1で至るところで微分不可能である

### 8.4 長さ無限と2次変分有限

**ブラウン運動の性質3**  
ブラウン運動の経路は確率1で有界変分ではない

標準ブラウン運動の場合を扱います。まずX(t)の区間[0,t]での変分を考えましょう。この区間に刻みを入れて

$0=t_{0}<t_{1}<\cdots < t_{n}=t$

と時間の分点をつくり（この分点系を#としておきます)、上昇・下降幅の合計(1次変分あるいは変分)

$V_{\#}=|X(t_{1})-X(t_{0})|+|X(t_{2})-X(t_{1})|+\cdots+|X(t_{n})-X(t_{n-1})|$

を計算すると、V#は分点の作り方#について有界ではありません。すなわち、いくらnを大きくして分点系#を細かくしても、時間区間の中での変化が激しいため、全体としてのVがいくらでも大きい場合が出てしまいます。ブラウン運動の経路は超細かくたたまれた長さ無限大の曲線なのです。

しかし、**2次変分**

$Q_{\#}=|X(t_{1})-X(t_{0})|^{2}+|X(t_{2})-X(t_{1})|^{2}+\cdots+|X(t_{n})-X(t_{n-1})|^{2}$

は有限なだけではなく、#を細かくすると、一定数tに**2次の平均収束**をします。

$E((Q_{\#}-t)^{2})\rightarrow 0 (||\#||\rightarrow 0)$

ここで$||\#||=max_{k}(t_{k}-t_{k-1})$で、分点系の細かさです。

証明は、

$t=\sum^{n}_{k=1}(t_{k}-t_{k-1})$

だから、節8.2の(iii)で$\mu=0$、したがって

$t_{k}-t_{k-1}=V(X(t_{k})-X(t_{k-1}))=E(X(t_{k})-X(t_{k-1}))^{2}$  

に注意して、

$Q_{\#}-t=\sum_{k=1}^{n}[(X(t_{k})-X(t_{k-1}))^{2}-(t_{k}-t_{k-1})]\\
=\sum_{k=1}^{n}[(X(t_{k})-X(t_{k-1}))^{2}-E(X(t_{k})-X(t_{k-1}))^{2}]$

とすると、式(2.4.5)を用いて、分散の加法性より

$E((Q_{\#}-t)^{2})=\sum^{n}_{k=1}E[(X(t_{k})-X(t_{k-1}))^{2}-(t_{k}-t_{k-1})]^{2}$

となります。(**こうなる理由が分からない**)ここで、

$Z_{k}^{2}=\frac{(X(t_{k})-X(t_{k-1}))^{2}}{t_{k}-t_{k-1}}$

はそれぞれ自由度1のχ2分布に従うので、代表的にそれを$Z^{2}$とおくと

$E((Q_{\#}-t)^{2})=E((Z^{2}-1)^{2})・\sum^{n}_{k=1}(t_{k}-t_{k-1})^{2}
\geq E((Z^{2}-1)^{2})・||\#||・t \rightarrow 0$
