### 位相空間
#### 空間とは何か
ある集合に属する元同士について何らかの関係が成り立つ、すなわち構造をもっているとき、空間と呼びます。例えば、ベクトル空間といったとき、その要素であるベクトルの間には和、定数倍といった計算が成り立ちます。

#### 位相とは何か
**定義 (開集合系・位相・位相空間)**　　
集合Sに対し、部分集合の族${\cal O}$がSの開集合系であるとは、次の条件1~3を満たすときをいいます。

1. $S\in {\cal O}$かつ$\varnothing \in {\cal O}$
2. $m \in \mathbb{N}, O_{1}, \cdots , O_{m} \in {\cal O} \Rightarrow O_{1} \cap \cdots \cap O_{m} \in {\cal O}$
3. 任意の集合$\wedge$にたいし、各元$\lambda \in \wedge$から${\cal O}$の元$O_{\lambda}\in {\cal O}$への対応を与えた時、$\cup_{\lambda \in \wedge}O_{\lambda} \in {\cal O}$

集合Sに開集合系$\cal O$が与えられているとき、「$\cal O$はSに位相を定める」もしくは「Sには$\cal O$による位相が入る」といい、$\cal O$の元を開集合と呼びます。このような位相構造が定められた集合Sを位相空間といいます。


開集合系はグループ分けとして捉えることができます。例えば、国家のグループ分けを考えます。集合S={英、仏、米}にたいし、次の部分集合族$\cal O$のうち、$(S,{\cal O})$が位相空間となるものはどれでしょうか？

1. {Φ,{英,仏},{仏,米},S}
2. {Φ,{英},{仏},{米},S}
3. {Φ,{米},{英,米},{仏,米},S}



### 1.2 Wiener空間
T>0を固定し、$W_{T}=W_{T}(\mathbb{R}^{d})$を[0,T]上の$w(0)=0$である$\mathbb{R}^{d}$値連続関数全体とします：

$W_{T}=\{w:[0,T] \rightarrow \mathbb{R}^{d} ;w$は連続かつ$w(0)=0\}$

$W_{T}$には一様収束の位相をいれます。この位相による$W_{T}$上のBorel σ加法族を${\cal B}(W_{T})$と書きます。次のWienerによる定理が現代確率論の出発点の１つです。

**定理1.2.1**  
$(W_{T},\cal{B}(W_{T}))$上の確率測度$\mu_{T}$で、$0=t_{0}<t_{1}<\cdots<t_{n}$と$A_{1},\cdots,A_{n}\in {\cal B}(\mathbb{R}^{d})$に対して、次をみたすものがただひとつ存在します。

$\mu_{T}(\{w \in W_{T}; w(t_{1}) \in A_{1} , \cdots ,w(t_{n})\in A_{n} \}\\
=\int_{A_{1}} dx_{1} \cdots \int_{A_{n}} dx_{n} \prod^{n}_{j=1}g(t_{j} -t_{j-1}, x_{j}, x_{j-1})$

ただし、$x_{0}=0$であり、$g(t,x,y)$は次で与えられます。

$g(t,x,y)=\frac{1}{(2\pi t)^{\frac{d}{2}}}  \exp(-\frac{|y-x|^{2}}{2t})$


定理1.2.1において与えられた$W_{T}$上の確率測度$\mu_{T}$をd次元Wiener測度とよび、確率空間$(W_{T},{\cal B}(W_{T}), \mu_{T})$を[0,T]上のd次元Wiener空間といいます。

**Cameron-Martin部分空間**

$\langle h_{1},h_{2}\rangle _{H_{T}}=\int^{T}_{0}\langle \dot{h_{1}}\rangle dt$

### 1.3 フィルター付き確率空間、適合確率過程
**定義1.3.3**  
フィルター付き確率空間$(\Omega, {\cal F}, P, \{{\cal F}_{t}\})$上で定義された$[0,\infty]-$値確率変数τに対して、

$\{\omega; \tau (\omega) \leq t \}\in {\cal F}_{t}$

がすべての$t \geq 0$に対して成り立つとき、τを$\{{\cal F}_{t}\}$に関する停止時刻といいます。

停止時刻とは、ある確率過程が何らかのふるまいを示したときにストップさせる時刻を表す確率変数です。

確率変数列$X_{1},X_{2}, \cdots $に対する停止時刻とは、確率変数$\tau \in \{1,2,\cdots \}$で、各時刻$t\in \{1,2,\cdots\}$において、あるイベントが起きる時刻がtであるという事象$\tau = t$は過去または現在$X_{1}, \cdots X_{t}$にのみ依存するような確率変数です。

例:赤か白かにかけるギャンブルに所持金100ドルで各ゲーム1ドルずつかけるギャンブラーを例に、停止時刻と停止ルールを具体的に考えます。

* 5回やったらやめる→停止ルールであり、P(τ=5)=1
* お金を使い果たすか、500回プレーしたらやめる→停止ルールである
* それ以上所持金が大きくなることはない最大値に到達したらやめる→未来の情報が必要であるため停止ルールではない。



### 1.4 離散時間マルチンゲール
**定義1.4.1**  
$X=\{X_{n}\}^{\infty}_{n=0}$を$\{{\cal F}_{n}\}$-適合、つまり各nに対して$X_{n}$が${\cal F}_{n}-$可測である実数値確率過程とします。このとき、各nに対して$X_{n}$が可積分で、

$E[X_{n+1}|{\cal F}_{n}]=X_{n}, \: P-a.s.$

が成り立つならば、Xを$\{{\cal F}_{n}\}$-マルチンゲールといいます。等号の代わりに、不等号$E[X_{n+1}|{\cal F}_{n}]\geq X_{n}, \: P-a.s.$が成り立つときXを劣マルチンゲール、逆の不等号が成り立つとき優マルチンゲールといいます。
